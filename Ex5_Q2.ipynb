```python
import numpy as np
```

```python
def find_nearest(array, value):
  idx = (np.abs(array-value)).argmin()
  return array[idx], idx
```

```python
def find_share(idx, res_a, res_b, player_a, player_b):
  up = res_b[idx] + (sum(res_b) - res_b[idx]) - (sum(res_a) - res_a[idx])
  down = player_a[idx] + player_b[idx]
  return up / down
```

```python
def reduce_shares(player_a, player_b, division):
  ratio = np.array(player_a / player_b)
  r, idx = find_nearest(ratio, 1.)

  for i in range(len(ratio)):
    if(ratio[i]>r):
      division[i] = 1
    else: 
      division[i] = 0

  res_a = player_a * division
  res_b = player_b * (1 - division)

  sum_a = sum(res_a)
  sum_b = sum(res_b)

  if (sum_a == sum_b):
    print("Sum of each player:", sum_a)
    return division

  else:
    x = find_share(idx, res_a, res_b, player_a, player_b)    
    division[idx] = x

    print("sum of each player:", sum(player_a * division))
    return division
```

```python
# Example 1

player_a = np.array([40, 30, 20, 10])
player_b = np.array([10, 20, 30, 40])
division = np.array([1., 0., 0.4, 0.7])

div = reduce_shares(player_a, player_b, np.copy(division))

print("Distribution to player A:",
      "\n\tBefore the improvement:", division, 
      "\n\tAfter the improvement:", div)
```
