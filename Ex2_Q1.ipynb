```python
import numpy as np
```

```python
# Create New Agent

class Agent():

    def __init__(self, optVal): 
      self.alloption = np.zeros(optVal.size)
      for i in range(optVal.size):
        self.alloption[i] = float(optVal[i])
        
    # INPUT: the index of an option(int)
    # OUTPUT: tha value(float) of the option to the aget
    def value(self, option):
      return self.alloption[option-1]
```

```python
# Creat Testing Data

Ami = Agent(np.array([1, 2, 3, 4, 5]))
Tmi = Agent(np.array([3, 1, 2, 5, 4]))
Rmi = Agent(np.array([3, 5, 5, 1, 1]))
```

```python
def isParetoImprovment(agents, option1, option2):  # Return bool
  for x in agents:
    if(x.value(option1) < x.value(option2)):
      return False 
  return True
```

```python
agents = [Ami, Tmi, Rmi]

print("\nOption 1 is Pareto Improvment:\n",
      isParetoImprovment(agents, 1, 2),
      isParetoImprovment(agents, 1, 3),
      isParetoImprovment(agents, 1, 4),
      isParetoImprovment(agents, 1, 5))

print("\nOption 2 is Pareto Improvment:\n",
      isParetoImprovment(agents, 2, 1),
      isParetoImprovment(agents, 2, 3),
      isParetoImprovment(agents, 2, 4),
      isParetoImprovment(agents, 2, 5))

print("\nOption 3 is Pareto Improvment:\n",
      isParetoImprovment(agents, 3, 1),
      isParetoImprovment(agents, 3, 2),
      isParetoImprovment(agents, 3, 4),
      isParetoImprovment(agents, 3, 5))

print("\nOption 4 is Pareto Improvment:\n",
      isParetoImprovment(agents, 4, 1),
      isParetoImprovment(agents, 4, 2),
      isParetoImprovment(agents, 4, 3),
      isParetoImprovment(agents, 4, 5))

print("\nOption 5 is Pareto Improvment:\n",
      isParetoImprovment(agents, 5, 1),
      isParetoImprovment(agents, 5, 2),
      isParetoImprovment(agents, 5, 3),
      isParetoImprovment(agents, 5, 4))
```

```python
def isParetoOptimal(agents, option, allOptions):  # Return bool
  for i in allOptions:
    if i == option:
      pass
    else:
      if isParetoImprovment(agents, option, i) == False:
        return False
  return True
```

```python
allOptions = [1, 2, 3, 4, 5]

print("Option 1 is Pareto Optimal:", 
      isParetoOptimal(agents, 1, allOptions))

print("Option 2 is Pareto Optimal:", 
      isParetoOptimal(agents, 2, allOptions))

print("Option 3 is Pareto Optimal:", 
      isParetoOptimal(agents, 3, allOptions))

print("Option 4 is Pareto Optimal:", 
      isParetoOptimal(agents, 4, allOptions))

print("Option 5 is Pareto Optimal:", 
      isParetoOptimal(agents, 5, allOptions))
```
