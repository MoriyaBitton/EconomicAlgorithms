```python
import networkx as nx

G = nx.DiGraph()
```

```python
def create_list(len):
  return [item for item in range(len)]
```

```python
def find_cycle_in_consumption_graph(allocation): #List[List[float]]

  players = len(allocation[:])    # Numbers of players = rows
  objects = len(allocation[:][0]) # Numbers of objects = cols

  # G is a Bilateral graph
  # 'players' side AND 'objects' side
  G.add_nodes_from(create_list(players + objects))

  for i in range(players):
    for j in range(objects):
      # There is Edge from player [i] to object [j]
      # if player [i] get more then 0% of object [j]
      if (allocation[i][j]>0):
        G.add_edge(i, players + j)
        
  try:
    cycle = list(nx.find_cycle(G, orientation="ignore"))
    print("There is cycle in the Graph:\n", cycle)
  except nx.NetworkXNoCycle:
    print("There is NO cycles in the Graph!")
```

```python
no_cycle = [[0.75, 0.0, 0.0],
            [0.25, 0.0, 0.0],
            [0.0, 0.5, 1.0]]

find_cycle_in_consumption_graph(no_cycle)
```

```python
one_cycle = [[0.75, 0.0, 0.0], 
             [0.25, 0.5, 1.0], 
             [0.0, 0.5, 1.0]]

find_cycle_in_consumption_graph(one_cycle)
```
